## 1) Oscilador armonico amortiguado y forzado por multiples frecuencias

Planteamos la ecuacion diferencial:
$$
m\ddot x + b\dot x + kx = F(t),
\qquad
F(t)=\sum_{i=1}^{n} f_i\cos(\gamma_i t).
$$
Definimos
$$
\omega_0^2=\frac{k}{m},\qquad 2\beta=\frac{b}{m}.
$$
Entonces:
$$
\ddot x +2\beta\dot x +\omega_0^2 x = \frac{1}{m}\sum_{i=1}^{n} f_i\cos(\gamma_i t).
$$

### 1.1 Solución exacta general para $x(t)$

La solución total es
$$
x(t)=x_h(t)+x_p(t).
$$

**Parte homogénea** ($\ddot x+2\beta\dot x+\omega_0^2x=0$):

Caso subamortiguado ($\beta<\omega_0$):
$$
x_h(t)=e^{-\beta t}\left(C_1\cos(\Omega t)+C_2\sin(\Omega t)\right),
\qquad \Omega=\sqrt{\omega_0^2-\beta^2}.
$$
(Si $\beta\ge \omega_0$, se reemplaza por las formas sobreamortiguada/crítica estándar).

**Parte particular** por superposición:
$$
x_p(t)=\sum_{i=1}^{n} x_i(t),
$$
donde cada $x_i$ satisface
$$
m\ddot x_i+b\dot x_i+kx_i=f_i\cos(\gamma_i t).
$$
Buscamos
$$
x_i(t)=A_i\cos(\gamma_i t)+B_i\sin(\gamma_i t).
$$
Sustituyendo y separando cosenos y senos:
$$
(k-m\gamma_i^2)A_i+b\gamma_i B_i=f_i,
$$
$$
(k-m\gamma_i^2)B_i-b\gamma_i A_i=0.
$$
Resolviendo:
$$
A_i=\frac{f_i(k-m\gamma_i^2)}{(k-m\gamma_i^2)^2+(b\gamma_i)^2},
\qquad
B_i=\frac{f_i b\gamma_i}{(k-m\gamma_i^2)^2+(b\gamma_i)^2}.
$$
Forma amplitud-fase equivalente:
$$
x_i(t)=\mathcal A_i\cos(\gamma_i t-\varphi_i),
$$
$$
\mathcal A_i=\frac{f_i}{\sqrt{(k-m\gamma_i^2)^2+(b\gamma_i)^2}},
\qquad
\tan\varphi_i=\frac{b\gamma_i}{k-m\gamma_i^2}.
$$

Por tanto, la **solución exacta** es:
$$
x(t)=e^{-\beta t}\left(C_1\cos(\Omega t)+C_2\sin(\Omega t)\right)
+\sum_{i=1}^{n}\frac{f_i}{\sqrt{(k-m\gamma_i^2)^2+(b\gamma_i)^2}}\cos(\gamma_i t-\varphi_i).
$$

### 1.2 Frecuencias igualmente espaciadas (peine): $\gamma_i=\gamma_0+(i-1)\Delta\gamma$

Al imponer
$$
\gamma_i=\gamma_0+(i-1)\Delta\gamma,\qquad i=1,\dots,n,
$$
la solucion forzada estacionaria queda
$$
x_{est}(t)=\sum_{i=1}^{n}\mathcal A(\gamma_i)\cos(\gamma_i t-\varphi(\gamma_i)),
$$
con
$$
\mathcal A(\gamma)=\frac{f(\gamma)}{\sqrt{(k-m\gamma^2)^2+(b\gamma)^2}},
\qquad
\varphi(\gamma)=\arctan\!\left(\frac{b\gamma}{k-m\gamma^2}\right).
$$

Si todos los tonos tienen misma amplitud $f_i=f_c$, entonces
$$
x_{est}(t)=\sum_{i=1}^{n}\frac{f_c}{\sqrt{(k-m\gamma_i^2)^2+(b\gamma_i)^2}}\cos(\gamma_i t-\varphi_i),
$$
que es la respuesta a un **peine de frecuencias** discretas.

### 1.3 Función de disipación de Rayleigh y potencia media absorbida

La fuerza viscosa es $F_d=-b\dot x$. Se escribe con función de disipación:
$$
\mathcal R=\frac12 b\dot x^2,
\qquad Q_d=-\frac{\partial \mathcal R}{\partial \dot x}=-b\dot x.
$$

En régimen estacionario, el promedio temporal de energía acumulada en el oscilador es constante, así que
$$
I=\langle P_{abs}\rangle = \langle F(t)\dot x(t)\rangle = b\langle \dot x^2\rangle.
$$
Para frecuencias distintas, los términos cruzados promedian a cero (promedio en ventana larga). Entonces:
$$
I=\sum_{i=1}^{n} I_i,
$$
$$
I_i=\frac12 b\gamma_i^2\mathcal A_i^2
=\frac12 b\gamma_i^2\frac{f_i^2}{(k-m\gamma_i^2)^2+(b\gamma_i)^2}.
$$
Por tanto
$$
I(\{\gamma_i\})=\frac12\sum_{i=1}^{n}\frac{b\gamma_i^2 f_i^2}{(k-m\gamma_i^2)^2+(b\gamma_i)^2}.
$$
Esta es la dependencia explícita de la potencia media absorbida con el conjunto de frecuencias aplicadas.

### 1.4 Uso del peine de frecuencias para estimar resonancia $\omega_0$

Con un peine de tonos $\gamma_i$ igualmente espaciados:

1. Se mide amplitud estacionaria $\mathcal A(\gamma_i)$ en cada línea del peine.
2. Se identifica el máximo de $\mathcal A$ (o de $I_i$).
3. Con amortiguamiento débil ($\beta\ll\omega_0$), la frecuencia de máximo de amplitud satisface
$$
\gamma_{max}\approx\sqrt{\omega_0^2-2\beta^2}\approx\omega_0.
$$

Método por fase:
- La fase teórica es
$$
\varphi(\gamma)=\arctan\!\left(\frac{b\gamma}{k-m\gamma^2}\right).
$$
- En el cruce de fase $\varphi=\pi/2$ se cumple $k-m\gamma^2=0$, luego
$$
\gamma=\omega_0.
$$

Ventaja del peine: se obtienen muchos puntos espectrales en una sola adquisición, permitiendo reconstruir rápidamente la curva de transferencia amplitud/fase alrededor de resonancia.

## 2) Circuito LC ideal y modulacion temporal de la capacitancia

Ley de voltajes de Kirchhoff:
$$
V_L+V_C=0,
$$
con
$$
V_L=L\frac{dI}{dt},\qquad V_C=\frac{q}{C},\qquad I=\dot q.
$$
Entonces
$$
L\ddot q+\frac{1}{C}q=0.
$$
Si $C$ es constante:
$$
\ddot q+\frac{1}{LC}q=0,
$$
y la frecuencia natural es
$$
\omega_0^2=\frac{1}{LC}.
$$

### 2.1 Capacitancia oscilatoria: reduccion a oscilador parametrico

Sea
$$
C(t)=C_0\big(1+h\cos(\gamma t)\big),\qquad h\ll1.
$$
La ecuacion exacta para $q$ es
$$
\ddot q+\frac{1}{L C(t)}q=0
=\ddot q+\frac{1}{L C_0(1+h\cos\gamma t)}q=0.
$$
Definiendo
$$
\omega_0^2=\frac{1}{L C_0},
$$
queda
$$
\ddot q+\frac{\omega_0^2}{1+h\cos\gamma t}q=0.
$$
Para $h\ll1$:
$$
\frac{1}{1+h\cos\gamma t}\approx 1-h\cos\gamma t+O(h^2),
$$
y por tanto
$$
\ddot q+\omega_0^2\big(1-h\cos\gamma t\big)q=0.
$$
Esta es una ecuacion de tipo Mathieu (oscilador parametrico).

### 2.2 Cercanía a $2\omega_0$: resonancia paramétrica

Tomamos
$$
\gamma=2\omega_0+\psi,
$$
con $|\psi|\ll\omega_0$.

La teoría de estabilidad de Mathieu (primer lóbulo de inestabilidad) da:
$$
|\gamma-2\omega_0|<\frac{h\omega_0}{2},
$$
equivalentemente
$$
|\psi|<\frac{h\omega_0}{2}.
$$

En esa región, la envolvente crece exponencialmente (inestabilidad paramétrica). Para detuning pequeño, la tasa de crecimiento aproximada es
$$
s \approx \frac12\sqrt{\left(\frac{h\omega_0}{2}\right)^2-\psi^2}.
$$
En resonancia exacta ($\psi=0$):
$$
s_{max}\approx\frac{h\omega_0}{4}.
$$

Conclusión física: al modular paramétricamente $C(t)$ cerca de $2\omega_0$, el sistema no absorbe energía por fuerza aditiva externa, sino por modulación del parámetro, produciendo amplificación de oscilaciones si se cumple la condición de inestabilidad anterior.

## Resumen final de resultados simbolicos

1. Oscilador amortiguado-forzado multifrecuencia:
$$
x(t)=x_h(t)+\sum_{i=1}^{n}\mathcal A_i\cos(\gamma_i t-\varphi_i),
$$
con
$$
\mathcal A_i=\frac{f_i}{\sqrt{(k-m\gamma_i^2)^2+(b\gamma_i)^2}},
\quad
\tan\varphi_i=\frac{b\gamma_i}{k-m\gamma_i^2}.
$$
2. Potencia media absorbida:
$$
I=\frac12\sum_{i=1}^{n}\frac{b\gamma_i^2 f_i^2}{(k-m\gamma_i^2)^2+(b\gamma_i)^2}.
$$
3. LC con $C(t)=C_0(1+h\cos\gamma t)$:
$$
\ddot q+\frac{1}{L C(t)}q=0,\quad \omega_0^2=\frac{1}{L C_0},\quad
\ddot q+\omega_0^2(1-h\cos\gamma t)q=0\; (h\ll1).
$$
4. Resonancia parametrica principal cerca de $\gamma=2\omega_0$:
$$
|\gamma-2\omega_0|<\frac{h\omega_0}{2}.
$$
