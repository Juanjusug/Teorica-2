## Enunciado A (imagen: Copia de Parcial 1 (1).png)

Considere un anillo de radio $R$ y masa $M$ que se cuelga de uno de sus puntos y oscila en su propio plano. Se comporta como un péndulo físico cuyo centro de gravedad está ubicado a una distancia $R$ del punto de suspensión. En el anillo se coloca una masa puntual $M$ que se puede deslizar sin fricción. Hallar el lagrangiano del sistema. Realice la aproximación de pequeñas oscilaciones. Resuelva el problema de vectores propios y valores propios implicado. Analice los resultados.

## A.1 Coordenadas generalizadas y lagrangiano

Tomamos dos coordenadas:
- $\theta$: ángulo del diámetro pivote-centro respecto a la vertical hacia abajo.
- $\phi$: posición angular de la masa deslizante sobre el anillo medida desde el radio centro$\to$pivote (en el marco del anillo).

Con el pivote como origen:
$$
\mathbf r_c = R(\sin\theta,\,\cos\theta),
\quad
\mathbf r_m = R(\sin\theta-\sin(\theta+\phi),\,\cos\theta-\cos(\theta+\phi)).
$$

Energía cinética:
$$
T = M R^2\dot\theta^2 + \frac{M R^2}{2}\left[\dot\theta^2+(\dot\theta+\dot\phi)^2-2\dot\theta(\dot\theta+\dot\phi)\cos\phi\right].
$$

Energía potencial (eje vertical positivo hacia abajo):
$$
U = -MgR\cos\theta - MgR\big(\cos\theta-\cos(\theta+\phi)\big)
= -MgR\big(2\cos\theta-\cos(\theta+\phi)\big).
$$

Lagrangiano: $L=T-U$.

## A.2 Pequeñas oscilaciones alrededor del equilibrio

Equilibrio estable: $\theta=0$, $\phi=\pi$ (masa deslizante en el punto más bajo del anillo).
Definimos $\eta=\phi-\pi$, con $|\theta|,|\eta|\ll 1$.

A segundo orden:
$$
T \simeq 3MR^2\dot\theta^2 + 2MR^2\dot\theta\dot\eta + \frac12 MR^2\dot\eta^2
= \frac12\begin{bmatrix}\dot\theta & \dot\eta\end{bmatrix}
\Big(MR^2\begin{bmatrix}6&2\\2&1\end{bmatrix}\Big)
\begin{bmatrix}\dot\theta\\\dot\eta\end{bmatrix}.
$$

$$
U \simeq \frac12\begin{bmatrix}\theta & \eta\end{bmatrix}
\Big(MgR\begin{bmatrix}3&1\\1&1\end{bmatrix}\Big)
\begin{bmatrix}\theta\\\eta\end{bmatrix}.
$$

El problema normal es:
$$
\det(\mathbf K-\omega^2\mathbf M_q)=0,
\quad
\mathbf M_q=MR^2\begin{bmatrix}6&2\\2&1\end{bmatrix},
\quad
\mathbf K=MgR\begin{bmatrix}3&1\\1&1\end{bmatrix}.
$$

```python
import sympy as sp

g, R, M = sp.symbols('g R M', positive=True)
w2 = sp.symbols('w2')

Mq = M*R**2*sp.Matrix([[6, 2], [2, 1]])
K = M*g*R*sp.Matrix([[3, 1], [1, 1]])

char_poly = sp.expand((K - w2*Mq).det())
roots = sp.solve(sp.Eq(char_poly, 0), w2)
char_poly, roots
```

Del cálculo simbólico:
$$
\omega_1^2=\frac{2g}{R},\qquad \omega_2^2=\frac{g}{2R}.
$$

Autovectores (modos):
- Para $\omega_1^2=2g/R$: $\eta=-3\theta$ (modo con fuerte movimiento relativo de la masa).
- Para $\omega_2^2=g/(2R)$: $\eta=0$ (modo casi rígido del conjunto alrededor del pivote).

## B.1 Esfera rodando: ligaduras en ángulos de Euler y no holonomía

Para esfera de radio $a$ rodando sin deslizamiento en el plano $(x,y)$:
$$
\dot x = a\,\omega_y,\qquad \dot y = -a\,\omega_x.
$$
Usando ángulos de Euler $(\phi,\theta,\psi)$ (convención $ZXZ$):
$$
\omega_x=\dot\phi\sin\theta\sin\psi + \dot\theta\cos\psi,
\quad
\omega_y=\dot\phi\sin\theta\cos\psi - \dot\theta\sin\psi.
$$
Entonces:
$$
\dot x-a(\dot\phi\sin\theta\cos\psi-\dot\theta\sin\psi)=0,
$$
$$
\dot y+a(\dot\phi\sin\theta\sin\psi+\dot\theta\cos\psi)=0.
$$

Forma Pfaff:
$$
\alpha_1=dx-a(\sin\theta\cos\psi\,d\phi-\sin\psi\,d\theta)=0,
$$
$$
\alpha_2=dy+a(\sin\theta\sin\psi\,d\phi+\cos\psi\,d\theta)=0.
$$

No son holónomas porque no pueden escribirse como $dF_1=0$, $dF_2=0$ globalmente; de hecho, sus derivadas exteriores contienen términos en $d\psi\wedge d\phi$ y $d\psi\wedge d\theta$ que no desaparecen y violan la condición de integrabilidad de Frobenius para un conjunto exacto.

## B.2 Rombo ($a$, $\pi/3$): tensor de inercia en un vértice

Tomamos el vértice $O$ como origen, con ejes en los lados del rombo:
$$
\mathbf e_1=(a,0),\qquad \mathbf e_2=(a/2,\sqrt3 a/2).
$$
Con densidad superficial uniforme $\sigma$ y masa total
$$
m=\sigma a^2\sin(\pi/3)=\sigma a^2\frac{\sqrt3}{2}.
$$

```python
u, v, a, sigma = sp.symbols('u v a sigma', positive=True)

x = a*(u + v/sp.Integer(2))
y = a*(sp.sqrt(3)/2)*v
J = a**2*sp.sqrt(3)/2  # jacobiano de la parametrizacion
dm = sigma*J

Ixx = sp.integrate(sp.integrate(y**2*dm, (u, 0, 1)), (v, 0, 1))
Iyy = sp.integrate(sp.integrate(x**2*dm, (u, 0, 1)), (v, 0, 1))
Ixy = -sp.integrate(sp.integrate(x*y*dm, (u, 0, 1)), (v, 0, 1))
Izz = sp.simplify(Ixx + Iyy)

m = sigma*a**2*sp.sqrt(3)/2
Ixx_m = sp.simplify(Ixx.subs(sigma, 2*m/(a**2*sp.sqrt(3))))
Iyy_m = sp.simplify(Iyy.subs(sigma, 2*m/(a**2*sp.sqrt(3))))
Ixy_m = sp.simplify(Ixy.subs(sigma, 2*m/(a**2*sp.sqrt(3))))
Izz_m = sp.simplify(Izz.subs(sigma, 2*m/(a**2*sp.sqrt(3))))

Ixx_m, Iyy_m, Ixy_m, Izz_m
```

```python
m, a = sp.symbols('m a', positive=True)
I2 = sp.Matrix([[sp.Rational(1,4)*m*a**2, -sp.Rational(5,8)/sp.sqrt(3)*m*a**2],
                [-sp.Rational(5,8)/sp.sqrt(3)*m*a**2, sp.Rational(2,3)*m*a**2]])

eigs = I2.eigenvects()
eigs
```

Resultados:
$$
I_{xx}=\frac14 ma^2,\qquad I_{yy}=\frac23 ma^2,\qquad I_{xy}=-\frac{5}{8\sqrt3}ma^2,\qquad I_{zz}=\frac{11}{12}ma^2.
$$
Momentos principales en el plano:
$$
I_1=\frac{7}{8}ma^2,\qquad I_2=\frac{1}{24}ma^2,
$$
y el tercero (normal al plano):
$$
I_3=I_{zz}=\frac{11}{12}ma^2.
$$
El ángulo de rotación de ejes principales satisface
$$
\tan(2\alpha)=\frac{2I_{xy}}{I_{xx}-I_{yy}}=\sqrt3 \Rightarrow 2\alpha=\frac{\pi}{3},\;\alpha=\frac{\pi}{6}.
$$

## B.3 Geometría modificada para cuerpo simétrico y ecuaciones de Euler

Una modificación simple: reemplazar el rombo por una placa circular (o cualquier placa con simetría de revolución alrededor del eje normal), de modo que
$$
I_1=I_2=I,\qquad I_3\neq I.
$$
Sin torque externo, las ecuaciones de Euler son:
$$
I\dot\omega_1+(I_3-I)\omega_2\omega_3=0,
$$
$$
I\dot\omega_2+(I-I_3)\omega_3\omega_1=0,
$$
$$
I_3\dot\omega_3=0.
$$
Luego $\omega_3=\Omega_3$ constante y, definiendo $\Omega=\dfrac{I_3-I}{I}\Omega_3$:
$$
\omega_1(t)=A\cos(\Omega t+\delta),\qquad \omega_2(t)=A\sin(\Omega t+\delta).
$$
Es decir, la proyección transversal de $\boldsymbol\omega$ precesa uniformemente alrededor del eje de simetría.
