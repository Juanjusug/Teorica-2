## 1) Ideas mas importantes

1. En pequenas oscilaciones, el sistema se linealiza alrededor del equilibrio.
2. La dinamica toma la forma matricial: $M\ddot{\eta}+K\eta=0$.
3. Los modos normales se obtienen con el problema de autovalores: $(K-\omega^2 M)a=0$.
4. Las frecuencias propias salen de la ecuacion secular: $\det(K-\omega^2 M)=0$.
5. Con coordenadas normales se desacopla el sistema en osciladores independientes.
6. Si hay degeneracion, una misma frecuencia tiene varios modos linealmente independientes.
7. La cuerda discreta (n masas) lleva en el limite continuo a la ecuacion de onda 1D.

## 2) Ecuaciones clave y cuando usarlas

### 2.1 Lagrangiano cuadratico
$$
T=\frac12\dot{\eta}^T M\dot{\eta},\qquad V=\frac12\eta^T K\eta,\qquad L=T-V
$$
Uso: punto de partida para pequenas perturbaciones alrededor del equilibrio.

### 2.2 Ecuaciones de movimiento acopladas
$$
M\ddot{\eta}+K\eta=0
$$
Uso: descripcion temporal completa del sistema linealizado.

### 2.3 Ansatz modal y problema secular
$$
\eta(t)=a\sin(\omega t+\phi),\quad (K-\omega^2 M)a=0,\quad \det(K-\omega^2M)=0
$$
Uso: hallar frecuencias propias y modos normales.

### 2.4 Ortogonalidad (metrica de masa)
$$
(u,v)=u^T M v,\qquad a_\alpha^T M a_\beta=\delta_{\alpha\beta}
$$
Uso: normalizar modos y proyectar condiciones iniciales.

### 2.5 Coordenadas normales
$$
Q=A^T\eta,\qquad AMA^T=I,\qquad AKA^T=\Omega^2
$$
Uso: desacoplar ecuaciones en la base modal.

### 2.6 Ecuaciones desacopladas
$$
\ddot Q_\alpha+\omega_\alpha^2Q_\alpha=0,\qquad Q_\alpha=C_\alpha\sin(\omega_\alpha t+\phi_\alpha)
$$
Uso: evolucion de cada modo normal por separado.

## 3) Casos del capitulo

### 3.1 Molecula triatomica lineal (longitudinal)
Frecuencias principales:
$$
\omega_1^2=0,\qquad \omega_2^2=\frac{k}{m},\qquad \omega_3^2=\frac{k}{m}\left(1+\frac{2m}{M}\right)
$$
- $\omega_1=0$: traslacion del centro de masa.
- $\omega_2,\omega_3$: vibraciones internas.

### 3.2 Potencial central $V(r)=A/r^n$
Condicion de estabilidad de orbita circular:
$$
\omega_r^2=(3-n)\Omega^2
$$
Estable si y solo si $n<3$.

### 3.3 Cuerda discreta y limite continuo
Cadena de $n$ masas (extremos fijos):
$$
\omega_\nu=2\sqrt{\frac{\kappa}{m}}\sin\left(\frac{\nu\pi}{2(n+1)}\right),\quad \nu=1,\dots,n
$$
Limite continuo (cuerda uniforme):
$$
\omega_\nu=\frac{\nu\pi}{L}\sqrt{\frac{\tau}{\mu}},\qquad
\frac{\partial^2\psi}{\partial t^2}=c^2\frac{\partial^2\psi}{\partial x^2},\ c^2=\frac{\tau}{\mu}
$$

```python
# Ejemplo simbolico 1: molecula triatomica lineal (longitudinal)
import sympy as sp

m, M, k, w2 = sp.symbols('m M k w2', positive=True)
K = k*sp.Matrix([[1,-1,0],[-1,2,-1],[0,-1,1]])
Mass = sp.diag(m, M, m)

# Ecuacion secular: det(K - w^2 M)=0
poly = sp.expand((K - w2*Mass).det())
sp.factor(poly)
```

```python
# Frecuencias propias (w^2) del ejemplo 1
sp.solve(sp.Eq(poly, 0), w2)
```

```python
# Ejemplo simbolico 2: estabilidad en V(r)=A/r^n
n, Omega = sp.symbols('n Omega', real=True)
omega_r2 = (3 - n)*Omega**2
omega_r2
```

Interpretacion del ejemplo 2:
- estable si $\omega_r^2>0\Rightarrow n<3$
- marginal si $n=3$
- inestable si $n>3$

```python
# Ejemplo simbolico 3: frecuencias de una cadena discreta
nu, N, kappa = sp.symbols('nu N kappa', integer=True, positive=True)
m_mass = sp.symbols('m_mass', positive=True)
omega_nu = 2*sp.sqrt(kappa/m_mass)*sp.sin(sp.pi*nu/(2*(N+1)))
sp.simplify(omega_nu)
```

## 4) Mini-guia de uso rapido

1. Define coordenadas pequenas alrededor del equilibrio.
2. Construye matrices $M$ y $K$.
3. Resuelve $\det(K-\omega^2M)=0$.
4. Halla autovectores y normaliza con $a_\alpha^TMa_\beta=\delta_{\alpha\beta}$.
5. Proyecta condiciones iniciales para obtener $C_\alpha,\phi_\alpha$.
6. Superpone modos para reconstruir $\eta(t)$.
