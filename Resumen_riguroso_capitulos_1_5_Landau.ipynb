## Capitulo 1. Movimiento de una particula

### 1) Espacio de configuracion y grados de libertad
- Un sistema mecanico se describe por coordenadas generalizadas $q_i$ ($i=1,\dots,s$).
- El numero $s$ de grados de libertad se obtiene restando restricciones holonomas independientes al numero de coordenadas cartesianas iniciales.
- El estado mecanico clasico instantaneo requiere $(q_i,\dot q_i)$ para todo $i$.

### 2) Transformaciones de coordenadas
- Si el cambio $q_i \mapsto q'_i(q,t)$ es regular (jacobiano no nulo), la descripcion es equivalente.
- La energia cinetica en coordenadas generalizadas tiene forma cuadratica en velocidades:
$$T(q,\dot q,t)=\frac12\sum_{i,j} a_{ij}(q,t)\dot q_i\dot q_j+\sum_i b_i(q,t)\dot q_i+c(q,t).$$

### 3) Fuerzas conservativas y potencial
- Si las fuerzas derivan de un potencial $U(q,t)$, entonces la dinamica se codifica mediante
$$L(q,\dot q,t)=T-U.$$
- Para potencial sin dependencia explicita en $t$, el sistema es autonomo y aparece una integral de energia.

### 4) Espacio de fases (idea previa)
- La evolucion mecanica se puede ver como una curva en un espacio de dimension $2s$.
- Esta observacion prepara la formulacion hamiltoniana posterior.

### 5) Puntos finos de examen
- Diferenciar claramente restriccion holonoma/no holonoma.
- Entender por que no toda fuerza admite potencial escalar global.
- No confundir coordenadas generalizadas con coordenadas cartesianas fisicas.

## Capitulo 2. Principio de minima accion

### 1) Funcional de accion
$$S[q]=\int_{t_1}^{t_2} L(q,\dot q,t)\,dt.$$
La trayectoria real hace estacionaria la accion: $\delta S=0$ bajo variaciones $\delta q_i(t)$ con extremos fijos ($\delta q_i(t_1)=\delta q_i(t_2)=0$).

### 2) Derivacion de Euler-Lagrange
Partiendo de
$$\delta S=\int_{t_1}^{t_2}\sum_i\left(\frac{\partial L}{\partial q_i}\delta q_i+\frac{\partial L}{\partial \dot q_i}\delta \dot q_i\right)dt,$$
integrando por partes el segundo termino y usando extremos fijos:
$$\delta S=\int_{t_1}^{t_2}\sum_i\left[\frac{\partial L}{\partial q_i}-\frac{d}{dt}\left(\frac{\partial L}{\partial \dot q_i}\right)\right]\delta q_i\,dt.$$
Como las variaciones son independientes:
$$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot q_i}\right)-\frac{\partial L}{\partial q_i}=0. $$

### 3) No unicidad del lagrangiano
- Si $L' = L + \frac{dF(q,t)}{dt}$, las ecuaciones de movimiento no cambian.
- En examen: justificar por que el termino total solo aporta borde en la accion.

### 4) Momento generalizado
$$p_i=\frac{\partial L}{\partial \dot q_i}. $$
No siempre coincide con $m\dot x_i$ en coordenadas no cartesianas o con potenciales dependientes de velocidad.

### 5) Coordenada ciclica
Si $\partial L/\partial q_k=0$, entonces
$$\dot p_k=0 \Rightarrow p_k=\text{cte}. $$
Este criterio es una via directa para obtener integrales primeras.

## Capitulo 3. Leyes de conservacion

### 1) Simetria temporal -> energia
Si $\partial L/\partial t=0$, se conserva
$$E=\sum_i \dot q_i\frac{\partial L}{\partial \dot q_i}-L. $$
Para $L=T-U$ con $U$ independiente de velocidades: $E=T+U$.

### 2) Simetria traslacional -> momento lineal
- Invariancia bajo traslaciones espaciales implica conservacion de $\mathbf{P}$.
- En sistema de particulas: $\mathbf{P}=\sum_a m_a\mathbf{v}_a$.

### 3) Simetria rotacional -> momento angular
- Invariancia bajo rotaciones implica conservacion de $\mathbf{M}$.
- Para una particula: $\mathbf{M}=\mathbf{r}\times\mathbf{p}$.

### 4) Centro de masa y separacion del movimiento
Para sistemas aislados:
$$M\ddot{\mathbf{R}}=0,\quad \mathbf{R}=\frac{1}{M}\sum_a m_a\mathbf{r}_a.$$
La dinamica se separa en traslacion del CM + movimiento relativo.

### 5) Conexion conceptual (Noether en version mecanica)
- Cada simetria continua diferenciable del lagrangiano induce una cantidad conservada.
- En examen, no basta enunciar: hay que identificar la transformacion y verificar invariancia de $L$ (o cuasi-invariancia por derivada total).

## Capitulo 4. Integracion de las ecuaciones de movimiento

### 1) Reduccion por integrales primeras
- Estrategia general: usar coordenadas ciclicas + energia para reducir orden del sistema.
- Si se obtienen suficientes integrales independientes en involucion (en formulacion hamiltoniana), la integracion se vuelve cuadraturas.

### 2) Movimiento en campo central
Con $U=U(r)$:
- El movimiento es plano ($\mathbf{M}=\text{cte}$).
- En coordenadas polares planas:
$$L=\frac{m}{2}(\dot r^2+r^2\dot\varphi^2)-U(r).$$
- Momento angular conservado:
$$m r^2\dot\varphi=l. $$
- Ecuacion radial equivalente a 1D con potencial efectivo:
$$\frac{m}{2}\dot r^2+U_{\text{ef}}(r)=E,\quad U_{\text{ef}}(r)=U(r)+\frac{l^2}{2mr^2}. $$

### 3) Problema de Kepler (resultado estructural)
Para $U(r)=-k/r$:
- Orbitas conicas (elipse, parabola, hiperbola) segun energia.
- Ley de areas por conservacion de $l$.
- Tercera ley para orbitas cerradas elipticas: $T^2\propto a^3$.

### 4) Dispersion en campo central (idea)
- Se relaciona parametro de impacto con angulo de dispersion usando la ecuacion radial.
- Cuidado con convenciones de angulo y limite de energia alta.

### 5) Errores tipicos
- Omitir el termino centrifugo en $U_{\text{ef}}$.
- Tratar $l$ como vector variable en campo central (debe ser constante).
- Confundir condicion de orbita circular estable: $U'_{\text{ef}}(r_0)=0$ y $U''_{\text{ef}}(r_0)>0$.

## Capitulo 5. Pequenas oscilaciones

### 1) Linealizacion alrededor del equilibrio
Sea $q_i=q_i^{(0)}+\eta_i$ con $|\eta_i|\ll 1$.
En equilibrio: $\partial U/\partial q_i|_0=0$.
A segundo orden:
$$T=\frac12\sum_{i,j} a_{ij}\dot\eta_i\dot\eta_j,\quad U=U_0+\frac12\sum_{i,j} c_{ij}\eta_i\eta_j.$$
Matrices $A=(a_{ij})$ simetrica definida positiva y $C=(c_{ij})$ simetrica.

### 2) Ecuaciones lineales acopladas
$$\sum_j a_{ij}\ddot\eta_j+\sum_j c_{ij}\eta_j=0. $$
Buscando modos normales $\eta_i=b_i e^{i\omega t}$:
$$\sum_j(c_{ij}-\omega^2 a_{ij})b_j=0.$$
Condicion de existencia no trivial:
$$\det(C-\omega^2 A)=0.$$

### 3) Modos normales
- Cada raiz $\omega_\alpha^2>0$ define una frecuencia propia.
- En coordenadas normales, el sistema se desacopla en osciladores armonicos independientes:
$$Q_\alpha''+\omega_\alpha^2 Q_\alpha=0. $$

### 4) Estabilidad
- Equilibrio estable local: forma cuadratica de $U$ positiva definida.
- Si algun $\omega_\alpha^2<0$, hay inestabilidad exponencial.

### 5) Caso con coordenadas ciclicas
- Pueden aparecer modos de frecuencia cero (modos neutros) asociados a simetrias continuas.
- En examen: interpretar fisicamente $\omega=0$ (no confundir con inestabilidad).

### Problema 1. Pendulo simple

Se toma como coordenada generalizada el angulo $\theta$ medido desde la vertical hacia abajo. Longitud $\ell$, masa $m$.

#### (a) Lagrangiano

La rapidez del bob es $v=\ell\dot\theta$.

$$T=\frac12 m\ell^2\dot\theta^2,\qquad U=mg\ell(1-\cos\theta).$$

Entonces

$$L=T-U=\frac12 m\ell^2\dot\theta^2-mg\ell(1-\cos\theta).$$

(Equivalentemente se puede usar $U=-mg\ell\cos\theta$; ambos difieren por constante.)

#### (b) Ecuacion exacta de movimiento

Sin friccion:

$$\frac{d}{dt}\left(\frac{\partial L}{\partial\dot\theta}\right)-\frac{\partial L}{\partial\theta}=0
\Rightarrow m\ell^2\ddot\theta+mg\ell\sin\theta=0.$$

Dividiendo por $m\ell^2$:

$$\ddot\theta+\omega_0^2\sin\theta=0,\qquad \omega_0^2=\frac g\ell.$$

Con friccion viscosa lineal (torque disipativo $-b\dot\theta$):

$$m\ell^2\ddot\theta+b\dot\theta+mg\ell\sin\theta=0
\Rightarrow
\ddot\theta+\gamma\dot\theta+\omega_0^2\sin\theta=0,$$

con

$$\gamma=\frac{b}{m\ell^2}. $$

#### (c) Pequenas oscilaciones reteniendo el siguiente orden (no linealidad cubica)

**Paso 1: Expansion de Taylor del seno**

Para ángulos pequeños $|\theta|\ll1$, expandemos $\sin\theta$ en serie de Taylor alrededor de $\theta=0$:

$$\sin\theta = \theta - \frac{\theta^3}{3!} + \frac{\theta^5}{5!} - \cdots = \theta - \frac{\theta^3}{6} + O(\theta^5).$$

**Paso 2: Sustitucion en la ecuacion diferencial**

Partimos de la ecuación exacta con friccion:

$$\ddot\theta+\gamma\dot\theta+\omega_0^2\sin\theta=0.$$

Reemplazamos la expansion aproximada de $\sin\theta$:

$$\ddot\theta+\gamma\dot\theta+\omega_0^2\left(\theta - \frac{\theta^3}{6}\right)=0.$$

Reordenando:

$$\ddot\theta+\gamma\dot\theta+\omega_0^2\theta - \frac{\omega_0^2}{6}\theta^3=0.$$

**Paso 3: Identificacion del tipo de ecuacion**

Esta es una ecuacion de **Duffing amortiguada**:

$$\boxed{\ddot\theta+\gamma\dot\theta+\omega_0^2\theta - \frac{\omega_0^2}{6}\theta^3=0}.$$

El coeficiente de $\theta^3$ es **negativo**: $-\omega_0^2/6 < 0$, lo que indica que es una ecuacion tipo Duffing **ablandante** (softening). Esto significa que la fuerza restauradora es mas debil para amplitudes grandes comparada con la aproximacion lineal.

**Paso 4: Interpretacion**

- El termino $\ddot\theta+\gamma\dot\theta+\omega_0^2\theta$ es la parte lineal (amortiguada).
- El termino $-\frac{\omega_0^2}{6}\theta^3$ es la primera correccion no lineal (proporcional al cubo del desplazamiento).
- Esta ecuacion es exacta hasta ordenes de $O(\theta^5)$ y superiores.

#### (d) Solucion en aproximacion lineal

**Paso 1: Linealizacion (eliminar termino cubico)**

Si nos quedamos solo con los terminos hasta orden $\theta$ (despreciamos $\theta^3$), obtenemos la ecuacion diferencial lineal:

$$\ddot\theta+\gamma\dot\theta+\omega_0^2\theta=0.$$

Esta es un ecuacion diferencial lineal homogenea de segundo orden con coeficientes constantes.

**Paso 2: Ecuacion caracteristica**

Buscamos soluciones de la forma $\theta(t)=e^{rt}$. Sustituyendo:

$$r^2 e^{rt} + \gamma r e^{rt} + \omega_0^2 e^{rt} = 0.$$

Dividiendo por $e^{rt}\neq 0$:

$$r^2 + \gamma r + \omega_0^2 = 0.$$

**Paso 3: Raices de la ecuacion caracteristica**

Aplicamos la formula de segundo grado:

$$r = \frac{-\gamma \pm \sqrt{\gamma^2 - 4\omega_0^2}}{2}.$$

El discriminante es:

$$\Delta = \gamma^2 - 4\omega_0^2.$$

**Paso 4: Condicion de subcritical damping (subamortiguamiento)**

Si $\gamma < 2\omega_0$, entonces $\Delta < 0$ y obtenemos raices complejas conjugadas:

$$r = -\frac{\gamma}{2} \pm i\Omega,$$

donde

$$\Omega = \sqrt{\omega_0^2 - \frac{\gamma^2}{4}}.$$

Esta cantidad $\Omega$ es la **frecuencia de oscilacion amortiguada**.

**Paso 5: Solucion general en el regimen subcritical**

La solucion general de la ecuacion lineal es una combinacion lineal de $e^{r_1 t}$ y $e^{r_2 t}$:

$$\theta_{\text{lin}}(t) = e^{-\gamma t/2}\left(C_1 e^{i\Omega t} + C_2 e^{-i\Omega t}\right).$$

Usando la identidad de Euler $e^{i\Omega t} = \cos(\Omega t) + i\sin(\Omega t)$, y eligiendo constantes reales $A_0$ y $\phi$ tales que $C_1$ y $C_2$ sean complejos conjugados apropiados, se obtiene:

$$\boxed{\theta_{\text{lin}}(t) = A_0 e^{-\gamma t/2} \cos(\Omega t + \phi)},$$

donde:
- $A_0 e^{-\gamma t/2}$ es la **amplitud que decae exponencialmente** con tasa de decaimiento $\gamma/2$.
- $\Omega = \sqrt{\omega_0^2 - \frac{\gamma^2}{4}}$ es la **frecuencia de oscilacion amortiguada**.
- $\phi$ es constante de fase determinada por condiciones iniciales.

**Paso 6: Casos limites**

1. **Sin friccion** ($\gamma=0$): $\Omega = \omega_0$, y $\theta_{\text{lin}}(t) = A_0\cos(\omega_0 t + \phi)$ (oscilacion pura a frecuencia natural).

2. **Friccion pequena** ($\gamma \ll \omega_0$): $\Omega \approx \omega_0\left(1-\frac{\gamma^2}{8\omega_0^2}\right) \approx \omega_0$ (la frecuencia es casi la natural).

3. **Amortiguamiento critico** ($\gamma = 2\omega_0$): $\Omega = 0$ (no hay oscilacion, solo decaimiento).

4. **Sobreamortiguamiento** ($\gamma > 2\omega_0$): no hay oscilacion, el sistema vuelve al equilibrio exponencialmente.

### Problema 2. Oscilador armonico forzado unidimensional

Partimos de

$$m\ddot x+m\omega^2x=F(t).$$

Definimos

$$\xi=\dot x+i\omega x.$$

#### (a) Demostrar la forma de primer orden

Derivando:

$$\dot\xi=\ddot x+i\omega\dot x.$$

Entonces

$$\dot\xi-i\omega\xi
=\ddot x+i\omega\dot x-i\omega(\dot x+i\omega x)
=\ddot x+\omega^2x
=\frac{F(t)}{m}.$$

Por tanto

$$\boxed{\dot\xi-i\omega\xi=\frac{F(t)}{m}}.$$

#### (b) Demostrar la solucion integral

Ecuacion lineal de primer orden:

$$\dot\xi-i\omega\xi=\frac{F(t)}{m}.$$

Multiplicamos por factor integrante $e^{-i\omega t}$:

$$\frac{d}{dt}\Big(\xi e^{-i\omega t}\Big)=\frac{F(t)}{m}e^{-i\omega t}.$$

Integrando de $0$ a $t$:

$$\xi(t)e^{-i\omega t}-\xi(0)=\frac1m\int_0^tF(\tau)e^{-i\omega\tau}\,d\tau.$$

Finalmente:

$$\boxed{\xi(t)=e^{i\omega t}\left[\frac1m\int_0^tF(\tau)e^{-i\omega\tau}\,d\tau+\xi(0)\right]}. $$

#### (c) Aplicacion concreta

Tomemos $F(t)=F_0\cos(\nu t)$ con condiciones iniciales de reposo:

$$x(0)=0,\qquad \dot x(0)=0\Rightarrow \xi(0)=0.$$

De la formula anterior (o equivalentemente resolviendo en forma real), para $\nu\neq\omega$:

$$\boxed{x(t)=\frac{F_0}{m(\omega^2-\nu^2)}\left[\cos(\nu t)-\cos(\omega t)\right]}. $$

- El primer termino es respuesta forzada.
- El segundo es transitorio de la frecuencia natural.

En resonancia ideal ($\nu=\omega$):

$$\boxed{x(t)=\frac{F_0}{2m\omega}\,t\sin(\omega t)},$$

crecimiento lineal en amplitud (sin disipacion).

### Problema 3. Tres cargas positivas en linea

Cargas: $Q$ (izquierda), $q$ (centro), $Q$ (derecha), todas positivas. Las dos $Q$ tienen masa infinita y la $q$ masa $m$.

Tomemos equilibrio geometrico en

$$x_1^{(0)}=-a,\qquad x_2^{(0)}=0,\qquad x_3^{(0)}=a.$$

#### (a) Posicion de equilibrio

Como las $Q$ externas son iguales y simetricas, la fuerza neta sobre $q$ en el punto medio es nula. Por tanto

$$x_{2,eq}=\frac{x_1+x_3}{2}\;(=0\text{ en el sistema elegido}).$$

Sobre cada $Q$ existe fuerza repulsiva finita; con masa infinita su aceleracion es cero y se consideran fijas (equilibrio impuesto por el modelo).

#### (b) Ecuacion linealizada de movimiento

Sea $x$ la pequena desviacion de la carga central respecto del equilibrio: $x_2=x$.

Fuerza total sobre $q$:

$$F(x)=kQq\left[\frac{1}{(a+x)^2}-\frac{1}{(a-x)^2}\right],$$

con $k=1/(4\pi\varepsilon_0)$.

Expandimos para $|x|\ll a$:

$$\frac{1}{(a\pm x)^2}=\frac1{a^2}\left(1\mp\frac{2x}{a}+O(x^2)\right).$$

Luego

$$F(x)\approx -\frac{4kQq}{a^3}x.$$

La ecuacion linealizada es

$$m\ddot x+\frac{4kQq}{a^3}x=0.$$

#### (c) Solucion por metodo de osciladores (modos normales)

Con masas externas infinitas, sus desplazamientos son nulos y queda un solo modo dinamico:

$$\omega^2=\frac{4kQq}{ma^3},\qquad
x(t)=A\cos(\omega t)+B\sin(\omega t).$$

Si $x(0)=x_0$ y $\dot x(0)=v_0$:

$$x(t)=x_0\cos(\omega t)+\frac{v_0}{\omega}\sin(\omega t).$$

#### (d) Discusion

1. El equilibrio del centro es estable: la fuerza restauradora es proporcional a $-x$.
2. La frecuencia crece con $Qq$ y decrece como $a^{-3/2}$ y $m^{-1/2}$.
3. El problema completo de 3 gdl colapsa a 1 gdl por el limite de masa infinita de las cargas externas.
4. Si se permitiera movimiento de las externas (masas finitas), aparecerian modos acoplados simetricos/antisimetricos y una formulacion matricial de varios osciladores.
